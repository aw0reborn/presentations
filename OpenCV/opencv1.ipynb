now lets try to crop the image

```python
img1 = cv2.imread("adi.jpg")
plt.imshow(cv2.cvtColor(img1, cv2.COLOR_BGR2RGB))

plt.show()
```

```python
image = img1[50:300, 400:800]
```

```python
cv2.imwrite("croped_adi.png",image)
```

```python
image_croped = cv2.imread("croped_adi.png")
plt.imshow(cv2.cvtColor(image_croped, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
save_jerry = cv2.imread("saveme.jpg")
plt.imshow(cv2.cvtColor(save_jerry,cv2.COLOR_BGR2RGB))
plt.show()
```
